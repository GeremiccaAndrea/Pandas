# Esercizi classe

```python
import pandas as pd
```

```python
movies = pd.read_csv('https://gist.githubusercontent.com/tiangechen/b68782efa49a16edaf07dc2cdaa855ea/raw/0c794a9717f18b094eabab2cd6a6b9a226903577/movies.csv')
movies.head()
```

```python
#visualizzare le ultime 8 righe del dataframe
movies.tail(8)
```

```python
#visualizzare solo le colonne che contengono il titolo e il genere del film
movies[['Genre']]
```

```python
#visualizzare le righe del dataframe dalla 20 alla 30
movies.iloc[20:31]
```

```python
#visualizzare il titolo e l’anno dei film che si trovano nelle posizioni 23, 29 e 45
movies.iloc[[23,9,45]][['Film', 'Year']]
```

```python
#visualizzare i titoli dei film del genere “Animation”
movies[movies.Genre == 'Animation'][['Film']]
```

```python
#visualizzare i titoli dei film prodotti dalla “Disney” e dalla “20th Century Fox”
movies[(movies['Lead Studio'] == 'Disney') | (movies['Lead Studio'] == '20th Century Fox')][['Film']]
```

```python
#visualizzare i titoli dei film usciti dal 2005 al 2010
movies[movies['Genre'] [2005:2010]]
```

```python
#diminuire del 10% il valore delle recensioni presenti nel campo Audience score %
movies['Audience score %'] = movies['Audience score %'] - (movies['Audience score %'] /100 * 10)
movies.head()
```

```python
#diminuire dei 10 il valore delle recensioni presenti nel campo Audience score %
movies['Audience score %'].subtract(10)
```

# ESERCIZI CASA 

```python
video = pd.read_csv('https://gist.githubusercontent.com/tiangechen/b68782efa49a16edaf07dc2cdaa855ea/raw/0c794a9717f18b094eabab2cd6a6b9a226903577/movies.csv')
video.head()
```

```python
#mostrare le prime 7 righe 
video[0:7]

#oppure 
video.head(7)
```

```python
#visualizzare le colonne film e genere 
video[['Film', 'Genre']]

#oppure 
video.iloc[: , 0:2] # : ti visualizza tutte le colonne 
                    # 0:2 le prime 2 colonne 
```

```python
#visualizzare le ultime 5 righe 
video.tail()
```

```python
#visualizzare la somma degli audencie score %
sum(video['Audience score %'])

#oppure 
video['Audience score %'].sum() #metodo pandas
```

```python
#visualizzare la media dei rotten tomatoes % 
video['Rotten Tomatoes %'].mean()
```

```python
#visualizzare i nomi dei film che sono prodotti dalla Universal 
video[video['Lead Studio'] == 'Universal'][['Film']]
```

```python
#visualizzare i film e la casa produttrice che hanno un rotten tomatoes % superiore alla media 
video[video['Rotten Tomatoes %']> video['Rotten Tomatoes %'].mean()][['Film', 'Lead Studio']]
```

```python
#visualizzare il film con l'audience score maggiore
video[video['Audience score %'] == video['Audience score %'].max()][['Film']]
```

```python
#visualizzare i film nelle righe 30,35,39
video[(video.index == 30 ) | (video.index == 35) | (video.index == 39)][['Film']]
```

```python
#visualizzare il minimo della colonna Worldwide Gross
video['Worldwide Gross'].min()
```

```python
#aumentare del 50% la colonna Audencie score %
# uguale a x = x + 1
video['Audience score %'] = video['Audience score %'] + (video['Audience score %'] /100 * 50)
video.head()
```

# Ricerca stringhe

```python
#film in cui compare la parola porno
movies[movies['Film'].str.contains('Porno')]
```

```python
#visualizzare film che iniziano con la lettere 'T'
movies[movies['Film'].str.startswith('T')]
```

```python
#visualizzare film che terminano con la lettere 'ON'
movies[movies['Film'].str.endswith('on')]
```

```python
#visualizzare i titoli dei film in ordine alfabetico 
movies.sort_values(by=['Film'])
```

```python
#visualizzare i nomi dei film dalla z alla a 
movies.sort_values(by = ['Film'], ascending = False)
```

```python
#ordinare il genere e il film 
movies.sort_values(by = ['Genre', 'Film'])
```
